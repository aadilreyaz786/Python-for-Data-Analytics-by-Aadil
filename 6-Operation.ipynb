# Filing Nan Values

```python
import numpy as np
```

```python
df = pd.DataFrame({'Col1': [10,20,30,np.nan],
                   'Col2': [np.nan,33,44,55],
                   'Col3': ['abc', 'def', 'ghi', 'jkl']})
```

```python
df.head()
```

```python
df.fillna(40)
```

```python
data = {'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.pivot_table(values= "D", index = ['A','B'], columns = ['C'])
```

```python
Money = {"Name": ['Aadil', 'Aadil', 'Aadil', 'Equbal', 'Equbal', 'Equbal'],
         'Currency': ['Dollar', 'Dollar', 'Rupee', 'Rupee', 'Dollar', 'Rupee'],
         'Note': [50,100,50,100,50,100],
         'Number of Note':[1,3,2,6,4,8]}
```

```python
df = pd.DataFrame(Money)
df
```
