# Multi Index & Hierarchy

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df =pd.DataFrame(np.random.rand(6,2), index= hier_index, columns = ['A','B'])
```

```python
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.loc['G1'].iloc[1]    # can use loc or iloc as done
```

```python
df.index.names
```

```python
df.index.names = ['Group', 'Num']
```

```python
df
```

```python
df.xs('G1')
```

```python
df.xs(['G1',1])
```

```python
df.xs(1,level='Num')
```
