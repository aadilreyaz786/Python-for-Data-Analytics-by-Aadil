```python
import numpy as np
import pandas as pd
```

***create DataFrame***

```python
data = {'Company':['Genpact','Accenture','Cognizant','Cognizant','Genpact','Accenture'],
        'Person': ['Aadil','Prashant', 'Neha','Simran','Mayank','Hasan'],
        'Sales':[100,120,340,80,150,230]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
df.groupby('Company').mean()
```

```python
df.groupby('Company').sum()
```

```python
By_Comp = df.groupby('Company')  # we can asssign a variable to apply aggregate fn
```

```python
By_Comp.mean()
```

```python
By_Comp.std()
```

```python
By_Comp.min()
```

```python
By_Comp.max()
```
