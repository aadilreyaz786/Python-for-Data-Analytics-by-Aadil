```python
import numpy as np
import pandas as pd
```

# Creating Series

```python
labels = ['a', 'b', 'c']
list = [10,20,30]
arr = np.array([100,200,300])
d = {'a':10, 'b':20,'c':30}
```

```python
pd.Series(labels)
```

```python
pd.Series(list)
```

```python
pd.Series(data=list,index=labels)
```

```python
pd.Series(d)
```

```python
pd.Series(arr)
```

```python
pd.Series(['aadil', 'imran', 'iqbal'])
```

```python
pd.Series([sum, len, print])
```

# Series Using Index

```python
Ser1= pd.Series([10,20,30,40],index = ['Patna', 'Delhi', 'Mumbai', 'Kolkata'])
```

```python
Ser1
```

```python
Ser2 = pd.Series([10,20,50,40],index=['Patna','Delhi','Hyderabad','Kolkata'])
Ser2
```

```python
Ser1 + Ser2
```

```python
Ser1['Patna']
```

```python
Ser2['Kolkata']
```

```python
Ser2[['Patna','Hyderabad']]   # We can pass list of index if we want munltiple data
```

```python
Ser2[3]                  # We can get value using index number
```

```python
Ser1 -Ser2
```

```python
Ser1*Ser2
```
