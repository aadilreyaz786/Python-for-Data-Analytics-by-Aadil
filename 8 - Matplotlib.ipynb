```python
import matplotlib.pyplot as plt
```

```python
x = [2,4,8,10,12,14]
y = [1,5,9,5,4,3]
```

# Plot line graph

```python
plt.plot(x,y,color = 'Purple')
```

```python
plt.plot(x,y,color = 'Purple',linestyle = '--', linewidth = 1)
plt.title("Stock Price")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

# Multiple line graph

```python
#Price of two stocks in different timeframe
time = [10,11,12,13,14,15]
p1 = [100,112,109,114,117,113]
p2 = [55,53,52,54,57,59]
plt.plot(time,p1)
plt.plot(time,p2)
```

```python
# Adding Subplot                     (plt.subplot(row,column,first graph)
x = [2,4,8,10,12,14]
y1 = [1,5,9,5,4,3]
y2 = [12,10,4,2,6,8]
plt.subplot(1,2,1)
plt.plot(x,y1, color='red', linestyle='-')
plt.subplot(1,2,2)
plt.plot(x,y2, color='blue', linestyle='--')
```

```python
# Graph with 2 rows and 1 column
x = [2,4,8,10,12,14]
y1 = [1,5,9,5,4,3]
y2 = [12,10,4,2,6,8]
plt.subplot(2,1,1)
plt.plot(x,y1, color='red', linestyle='-')
plt.subplot(2,1,2)
plt.plot(x,y2, color='blue', linestyle='--')
```

# Bar Plot

```python
Covid_cases = [100, 115, 150, 200, 450, 350, 270, 120, 50]
time = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept']
plt.bar(time, Covid_cases)
plt.title('Covid Data')
plt.xlabel('Covid 2nd Phase')
plt.ylabel('Cases in Million')
plt.show()
```

```python
names = ['A', 'B', 'C', 'D']
weight = [55, 50, 60, 45]
plt.bar(names,weight, color = 'red')
plt.show()
```

```python
#Multiple Bar plots
names = ['A', 'B', 'C', 'D']
weight = [55, 50, 60, 45]
Age = [ 20,29,24,35]
plt.bar(names,weight, color="grey")
plt.bar(names,Age)
plt.show()
```

```python
#Subplot
plt.subplot(1,2,1)
plt.bar(names,weight)
plt.subplot(1,2,2)
plt.bar(names,Age)
plt.show()
```

```python
#Horizontal Bar Plot
names = ['A', 'B', 'C', 'D']
weight = [55, 50, 60, 45]
plt.barh(names,weight)
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
import numpy as np
import random
```

```python
x = np.arange(0,10)
y = np.random.randint(10,50,10)
```

```python
plt.scatter(x,y, color='blue', marker='*', s=100)
plt.show()
```

```python
# Addiding multiple Scatter plot
x = np.arange(0,10)
y = np.random.randint(10,50,10)
z = [5,7,13,8,15,25,23,14,19,38]
plt.scatter(x,y)
plt.scatter(x,z,color='red', marker='*',s=100)
plt.show()
```

# Box &Whisker Plot

```python
x = [5,3,4,2,6,8,10,9,12,15]
y = [1,2,3,4,5,6,7,8,9,17]
z = [4,9,3,8,10,7,5,2,1,12]
p = [x,y,z]
plt.boxplot(p)
plt.show()
```
